```python
# importing necessary packages
import pandas as pd
import numpy as np
```

```python
# code to load  predictions of SVD,NMF,CoClustering,KNN,Neural Net, ALS models into a dataframe
SVD_df = pd.read_csv('C:/Users/rvrma/Documents/3rd semester/Massive Data Mining/project/predictions/SVD.csv')
SVD_df.rename(columns = {'predicted_rating':'SVD_pred_rating'}, inplace = True)
NMF_df = pd.read_csv('C:/Users/rvrma/Documents/3rd semester/Massive Data Mining/project/predictions/NMF.csv')
NMF_df.rename(columns = {'predicted_rating':'NMF_pred_rating'}, inplace = True)
CoClustering_df = pd.read_csv('C:/Users/rvrma/Documents/3rd semester/Massive Data Mining/project/predictions/CoClustering.csv')
CoClustering_df.rename(columns = {'predicted_rating':'CoClustering_pred_rating'}, inplace = True)
KNN_df = pd.read_csv('C:/Users/rvrma/Documents/3rd semester/Massive Data Mining/project/predictions/KNNBaseline.csv')
KNN_df.rename(columns = {'predicted_rating':'KNN_pred_rating'}, inplace = True)
NN_df = pd.read_csv('C:/Users/rvrma/Documents/3rd semester/Massive Data Mining/project/predictions/NeuralNet_Prediction.csv')
NN_df.rename(columns = {'predicted_rating':'NN_pred_rating'}, inplace = True)
ALS_df = pd.read_csv('C:/Users/rvrma/Documents/3rd semester/Massive Data Mining/project/predictions/ALS_results.csv')
ALS_df.rename(columns = {'predicted_rating':'ALS_pred_rating'}, inplace = True)

# creating a dataframe for ensemble ratings which is the sum of all other models ratings
ensemble_dataframe = SVD_df
ensemble_dataframe = ensemble_dataframe[['userId','movieId','true_rating','SVD_pred_rating']]
ensemble_dataframe['NMF_pred_rating'] = NMF_df['NMF_pred_rating']
ensemble_dataframe['CoClustering_pred_rating'] = CoClustering_df['CoClustering_pred_rating']
ensemble_dataframe['KNN_pred_rating'] = KNN_df['KNN_pred_rating']
ensemble_dataframe['NN_pred_rating'] = NN_df['NN_pred_rating']
ensemble_dataframe['ALS_pred_rating'] = ALS_df['ALS_pred_rating']
#ensemble_dataframe['ALS_pred_rating'] = ensemble_dataframe['ALS_pred_rating'].fillna(0)

#finding ensemble predicted ratings
ensemble_dataframe['Ensemble_rating'] = ( ensemble_dataframe['SVD_pred_rating'] + ensemble_dataframe['NMF_pred_rating'] + ensemble_dataframe['CoClustering_pred_rating'] + ensemble_dataframe['KNN_pred_rating'] + ensemble_dataframe['NN_pred_rating'] + ensemble_dataframe['ALS_pred_rating'] )/6
ensemble_dataframe.head()
```

```python
# dropping all model ratings except ensemble rating
en_df = ensemble_dataframe.drop(['SVD_pred_rating','NMF_pred_rating','CoClustering_pred_rating','KNN_pred_rating','NN_pred_rating','ALS_pred_rating'],axis=1)
en_df = en_df[['userId','movieId','Ensemble_rating','true_rating']]
```

```python
en_df.rename({'Ensemble_rating':'predicted_rating'}, axis = 1, inplace = True)
en_df.head()
```

```python
# saving ensemble predicted ratings into a csv file 
en_df.to_csv('C:/Users/rvrma/Documents/3rd semester/Massive Data Mining/project/predictions/Ensemble.csv') 
```
